```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df1 =pd.read_csv('u.data')
df1
```

```python
column_names = ['user_id' , 'item_id', 'rating', 'timestamp']
```

```python
df = pd.read_csv('u.data' , sep='\t', names= column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
```

```python
movie_titles.head()
```

```python
df = pd.merge(df,movie_titles,how='inner',on='item_id')
```

```python
df.head(20)
```

```python
sns.set_style('white')
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending= False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending= False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
ratings.head()
```

```python
ratings['num'] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
plt.figure(figsize=(10,6))
ratings['num'].hist(bins=70)
```

```python
plt.figure(figsize=(10,6))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating', y='num', data= ratings, alpha=0.6)
```

```python
moviemat = df.pivot_table(index= 'user_id', columns= 'title', values= 'rating')
```

```python
moviemat.head()
```

```python
ratings.sort_values('num', ascending= False).head(10)
```

```python
star_wars_ratings = moviemat['Star Wars (1977)']
liar_liar_ratings = moviemat['Liar Liar (1997)']
```

```python
star_wars_ratings.head(10)
```

```python
similar_to_starwars = moviemat.corrwith(star_wars_ratings)
```

```python
similar_to_liarliar = moviemat.corrwith(liar_liar_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
corr_starwars.dropna(inplace = True)
```

```python
corr_starwars.head(10)
```

```python
corr_starwars.sort_values('Correlation', ascending= False). head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num'])
```

```python
corr_starwars.head(10)
```

```python
corr_starwars[corr_starwars['num']>60].sort_values('Correlation', ascending= False).head(10)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['Correlation'])
corr_liarliar.dropna(inplace = True)
```

```python
corr_liarliar.head()
```

```python
corr_liarliar = corr_liarliar.join(ratings['num'])
```

```python
corr_liarliar.head(5)
```

```python
corr_liarliar[corr_liarliar['num']>60].sort_values('Correlation', ascending = False).head()
```
